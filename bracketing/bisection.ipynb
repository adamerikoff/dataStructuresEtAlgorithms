```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bisection(func, a, b, tol=1e-6, max_iter=100):
    if func(a) * func(b) >= 0:
        raise ValueError("Function has the same sign at both interval endpoints. Bisection method cannot be applied.")

    c = (a + b) / 2  # Compute the midpoint of the interval

    for i in range(max_iter):
        c = (a + b) / 2  # Compute the midpoint of the interval

        if func(c) == 0 or (b - a) / 2 < tol:
            return c  # Found the root or the interval is small enough

        if func(c) * func(a) < 0:
            b = c  # Replace the right endpoint with the midpoint
        else:
            a = c  # Replace the left endpoint with the midpoint
    return (a + b) / 2  # Return the best estimate of the root found within the max number of iterations
```

```python
def unimodal_function(x):
    return (x - 3)**2 -3


x = np.linspace(-2, 8, 400)
y = unimodal_function(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x - 3)^2 + 4', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = (x - 3)^2 + 4')
root = bisection(unimodal_function, -2, 4)
plt.scatter(root, unimodal_function(root), color="red")
plt.grid(True)
plt.legend()
plt.show()
```
